```python
import pandas as pd
import numpy as np  
```

```python
df=pd.read_excel('British Airways Summer Schedule Dataset - Forage Data Science Task 1.xlsx')
```

```python
df.head()
```

```python
df['category']=df['HAUL']+' '+df['TIME_OF_DAY']
```

```python
df[['FLIGHT_TIME','HAUL','category']].head()
```

```python
lookup = pd.DataFrame({
    'category': [
        'SHORT Morning', 'SHORT Afternoon', 'SHORT Evening',
        'LONG Morning', 'LONG Afternoon', 'LONG Evening' ],
    'tier1_pct': [0.04, 0.03, 0.02, 0.10, 0.08, 0.06],
    'tier2_pct': [0.18, 0.15, 0.12, 0.35, 0.30, 0.25],
    'tier3_pct': [0.55, 0.50, 0.45, 0.80, 0.75, 0.70]
})
```

```python
df=df.merge(lookup,on='category',how='left')
```

```python
df.head()
```
